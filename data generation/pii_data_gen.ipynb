```python
import pandas as pd
```

```python
df = pd.read_json('100_dataset.jsonl', lines=True)
df
```

```python
df[['First Name', 'Last Name', 'ID', 'Unknown', 'Phone Number', 'Address', 'Email', 'Link']] = df['profile'].str.split(';', expand=True)
```

```python
df1 = df.drop(columns=['id', 'profile', 'Unknown'])
df1
```

```python
df1['Link'] = df['Link'].apply(lambda x: x[:-1] if x.endswith('/') else x)
df1
```

```python
import re

def tokenize_sentence(sentence):
    tokens = re.findall(r'\b(?:[^\s@]+@[^\s@]+\.[a-zA-Z]{2,}|https?://(?:[^\s/$.?#][^\s]*)+(?:/[^/\s]*)?/?)\b|\b\w+\b|[()\[\]{}]|[^\w\s()]', sentence)
    return tokens
```

```python
text = "This is an example text with an email@example.com and a link (http://example.com/)"

print(tokenize_sentence(text))
```

```python
df1['tokenized response'] = df1['response'].apply(tokenize_sentence)
```

```python
df1
```

```python
df1.columns
```

```python
def assign_labels(row):
    labels = []
    for token in row['tokenized response']:
        if row['First Name'] in token or row['First Name'] == token:
            labels.append('B-NAME_STUDENT')
        elif row['Last Name'] in token or row['Last Name'] == token:
            labels.append('I-NAME_STUDENT')
        elif row['ID'] in token or row['ID'] in token:
            labels.append('-ID_NUM')
        elif token == row['Phone Number'] or row['Phone Number'] in token:
            labels.append('B-PHONE_NUM')
        elif token == row['Email'] or row['Email'] in token:
            labels.append('B-EMAIL')
        elif token == row['Link'] or row['Link'] in token:
            labels.append('B-URL_PERSONAL')
        elif token == row['Address'].split()[0]:  # First word of address
            labels.append('B-STREET_ADDRESS')
        elif token in row['Address'].split()[1:]:  # Last words of address
            labels.append('I-STREET_ADDRESS')
        else:
            labels.append('O')
    return labels
```

```python
df1['labels'] = df1.apply(assign_labels, axis=1)
```

```python
final_df = df1[['tokenized response', 'labels']].copy()
final_df
```

```python
file_path = '100_gen_data.csv'
final_df.to_csv(file_path, index=False)
```
